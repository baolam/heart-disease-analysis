```python
import os
import json
import pandas as pd
```

```python
RAW_FOLDER = "raw_concat"
CSV_FOLDER = "data_csv_Thang"
```

```python
os.makedirs(CSV_FOLDER, exist_ok=True)
```

```python
data_names = set()
for file in os.listdir(RAW_FOLDER):
    name = file.split('.')[0]
    data_names.add(name)
# Lấy hết tên trong file raw_concat để làm tên file cho folder data_csv_Thang
```

```python
print(data_names)
```

```python
for name in data_names:
    with open(RAW_FOLDER + "/" + name + '.json', "r", encoding= "utf-8") as f:
        data = json.load(f)

    df = pd.DataFrame(data)
    df.to_csv(CSV_FOLDER + "/" + name + '.csv', index=False, encoding='utf-8')
```
