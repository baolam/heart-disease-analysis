```python
import pandas as pd
import requests 
import json
```

```python
def get_indicators(url):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()

        indicators = data.get('value', [])

        for i in indicators:
            print(i['IndicatorCode'] + ': ' + i['IndicatorName'])
    else:
        print('failed')
```

```python
topics = ['glucose', 'cholesterol', 'tobacco']
for topic in topics:
    print('TOPIC: ',topic)
    url = f"https://ghoapi.azureedge.net/api/Indicator?$filter=contains(IndicatorName,'{topic}')"
    get_indicators(url)
    print('/n')
```
