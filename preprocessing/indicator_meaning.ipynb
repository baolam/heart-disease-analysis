```python
import pandas as pd
import requests 
import json
import os
```

```python
def get_indicators(url):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()

        try:
            i = data.get('value', [])
            return (i['IndicatorCode'] + ': ' + i['IndicatorName'] + '\n')
        except:
            return []
    else:
        print('failed', response.status_code)
```

```python
topics = ['cardiovascular']
for topic in topics:
    print('TOPIC: ',topic)
    url = f"https://ghoapi.azureedge.net/api/Indicator?$filter=contains(IndicatorName,'{topic}')"
    get_indicators(url)
    print('/n')
```

```python
with open("../data/urls/cardiovascular_diseases.txt", "r", encoding="utf-8") as f:
    for indicator in f:
        indicator = indicator.strip()
        url = f"https://ghoapi.azureedge.net/api/Indicator?$filter=contains(IndicatorCode,'{indicator}'"
        get_indicators(url)
```
