```python
import pandas as pd
import requests 
import json
import os
```

```python
def get_indicators(url):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()

        try:
            i = data.get('value', [])
            return (i['IndicatorCode'] + ': ' + i['IndicatorName'] + '\n')
        except:
            return []
    else:
        print('failed')
        return []
```

```python
def get_meaning_indicator(indicator):
    url = f"https://ghoapi.azureedge.net/api/Indicator?$filter=contains(IndicatorCode,'{indicator}')"
    data = get_indicators(url)
    return data
```

```python
#topics = ['glucose', 'cholesterol', 'tobacco']
for file in os.listdir('../data/urls'):
    with open(f"../docs/{file}", 'w', encoding='utf-8') as f1:
        print("TOPIC: ", file.split('.')[0])
        with open ('../data/urls/'+file, 'r', encoding= 'utf-8' ) as f:
            for indicator in f:
                indicator = indicator.strip()
                mean_indicator=get_meaning_indicator(indicator)
                f1.write(mean_indicator + '\n')
```
