```python
import os
import pandas as pd
import json
```

```python
READ_PATH = "../data/raw_concat"
CSV_PATH = "../data/csv"
```

```python
os.makedirs(CSV_PATH, exist_ok=True)
```

```python
def export_csv_file(name="", export_as_name=""):
    fname = os.path.join(READ_PATH, name)
    with open(fname, "r", encoding="utf-8") as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    df.to_csv(os.path.join(CSV_PATH, export_as_name))
```

```python
# Thử ví dụ tạo file csv cho dữ liệu air
#export_csv_file("air_pollution.json", "air_pollution.v1.csv")
topics = ['cholesterol', 'glucose', 'tobacco']
for topic in topics:
    export_csv_file(f"{topic}.json", f"{topic}.csv")
```

```python
data_names = set()
for file in os.listdir(READ_PATH):
    name = file.split('.')[0]
    data_names.add(name)
for name in data_names:
    export_csv_file(name + ".json", name + ".csv" )
```
