```python
df = pd.read_csv("../../data/csv/infrastructure.csv")
```

```python
df["IndicatorCode"].unique()
```

```python
Unnecessary_indicatorcodes = [
    "NCD_HYP_CONTROL_A",
    "NCD_HYP_CONTROL_C",
    "NCD_HYP_PREVALENCE_A",
    "NCD_HYP_PREVALENCE_C"
]

df = df[~df["IndicatorCode"].isin(Unnecessary_indicatorcodes)]
df["IndicatorCode"].unique()
```

```python
df.drop(columns= "Value", axis = 1, errors= "ignore", inplace=True)
```

```python
print(df["ParentLocationCode"].isnull().value_counts())
print(df["SpatialDim"].isnull().value_counts())
print(df["NumericValue"].isnull().value_counts())
print(df["Date"].isnull().value_counts())
print(df["IndicatorCode"].isnull().value_counts())
```

```python
df["Date"] = pd.to_datetime(df["Date"], format= "ISO8601", utc = True)
```
