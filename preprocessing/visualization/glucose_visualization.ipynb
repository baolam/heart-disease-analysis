```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open ('../../data/raw_concat/glucose.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)
print(df.head())
```

```python
df.info()
df.describe()
```

```python
df['NumericValue'].isnull().sum()
```

```python
# remove null value
print(len(df))
df=df.dropna(subset=['NumericValue'])
print(len(df))
print(df["NumericValue"].isnull().sum())
```

```python
plt.figure()
plt.hist(df["NumericValue"], bins=20)
plt.xlabel("Glucose_Value")
plt.ylabel("Frequency")
plt.title("Distribution of Value")
plt.show()
```

```python
plt.figure()
df.boxplot(column="NumericValue", by="IndicatorCode")
plt.title("Value by Group")
plt.suptitle("")
plt.show()
```
