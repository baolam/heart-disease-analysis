```python
# Tổng hợp chương trình test toàn bộ
CONCAT_DATASET = "raw_concat"
```

```python
def count_countries():
    for index in os.listdir(CONCAT_DATASET):
        name = index.split('.')[0]
        with open(CONCAT_DATASET + '/' + index, "rb") as f:
            data = json.load(f)
        df = pd.DataFrame(data)
        countries = df["SpatialDim"].unique()
        times = pd.to_datetime(df["Date"], format="ISO8601", utc=True)
        length, len_countries = len(data), len(countries)
        min_date, max_date = times.min().year, times.max().year
        
        print(f"Bộ {name} có {length} mẫu, thu thập từ {len_countries} quốc gia, trong khoảng thời gian từ {min_date} đến {max_date}!")
count_countries()
```
